```python
### 计算图像数据集的均值和方差

import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
#from scipy.misc import imread
 
filepath = r'/home/moyan/chest_xray/train/PNEUMONIA/'  # 数据集目录
pathDir = os.listdir(filepath)
 
R_channel = 0
G_channel = 0
B_channel = 0
for idx in range(len(pathDir)):
    filename = pathDir[idx]
    img = cv2.imread(os.path.join(filepath, filename)) / 255.0
    R_channel = R_channel + np.sum(img[:, :, 0])
    G_channel = G_channel + np.sum(img[:, :, 1])
    B_channel = B_channel + np.sum(img[:, :, 2])
num = len(pathDir) * 1468* 1636 # 这里（512,512）是每幅图片的大小，所有图片尺寸都一样
R_mean = R_channel / num
G_mean = G_channel / num
B_mean = B_channel / num
 
R_channel = 0
G_channel = 0
B_channel = 0
for idx in range(len(pathDir)):
    filename = pathDir[idx]
    img = cv2.imread(os.path.join(filepath, filename)) / 255.0
    R_channel = R_channel + np.sum((img[:, :, 0] - R_mean) ** 2)
    G_channel = G_channel + np.sum((img[:, :, 1] - G_mean) ** 2)
    B_channel = B_channel + np.sum((img[:, :, 2] - B_mean) ** 2)
R_var = np.sqrt(R_channel / num)
G_var = np.sqrt(G_channel / num)
B_var = np.sqrt(B_channel / num)


print("R_mean is %f, G_mean is %f, B_mean is %f" % (R_mean, G_mean, B_mean))
print("R_var is %f, G_var is %f, B_var is %f" % (R_var, G_var, B_var))
```

```python
R_std = np.sqrt(R_var)
G_std = np.sqrt(G_var)
B_std = np.sqrt(G_var)
print("R_std is %f, G_std is %f, B_stdis %f" % (R_std, G_std, B_std))
```

```python
def im_convert(tensor):
    """展示数据"""
    
    image=tensor.to("cpu").clone().detach()#调用CPU
    image=image.numpy().squeeze()
    image=image.transpose(1,2,0)  #plt.imshow在现实的时候输入的是（imagesize,imagesize,channels）但是数据保存时（channels，imagesize,imagesize）
    image=image*np.array((0.523001, 0.523001, 0.523001))+np.array((0.496093, 0.496093, 0.496093)) #前面做了normalize [0.485, 0.456, 0.406], [0.229, 0.224, 0.225]
    return image
```

```python
fig=plt.figure(figsize=(20,12))
cloumns=5
rows=2

dataiter=iter(dataloaders_train)
inputs,classes =dataiter.next()

#依次写入图片
for idx in range (cloumns*rows):
    ax=fig.add_subplot(rows,cloumns,idx+1,xticks=[],yticks=[])
    ax.set_title(int(class_names[classes[idx]]))
    plt.imshow(im_convert(inputs[idx]))  #单张传入
plt.show()  #窗口展示
```
